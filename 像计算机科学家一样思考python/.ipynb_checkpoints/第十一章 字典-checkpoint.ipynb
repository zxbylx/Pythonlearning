```python
#使用字典作为计数器集合
def histogram(s):
    d=dict()
    for c in s:
        if c not in d:
            d[c]=1
        else:
            d[c]+=1
    return d
h=histogram('faljgaldkjalgkjeweraldfjsl')
print(h)
```

```python
#练习：使用get将histogram写得更紧凑些，你应当可以消除掉if语句
def histogram(s):
    d=dict()
    for c in s:
        d[c]=int(d.get(c,'0'))+1
    return d
h=histogram('faljgaldkjalgkjeweraldfjsl')
print(h)
```

```python
#反向查找，接收一个值，并返回映射到该值的第一个键
def reverse_lookup(d,v):
    for k in d:
        if d[k]==v:
            return k
    raise ValueError
    
h=histogram('parrot')
k=reverse_lookup(h,2)
print(k)
```

```python
#练习：修改reverse_lookup函数，使得它建立并返回一个包含所有映射到v的键的列表，没有找到返回空列表
def reverse_lookup(d,v):
    keys=[]
    for k in d:
        if d[k]==v:
            keys.append(k)
    return keys
h=histogram('aabbccddd')
reverse_lookup(h,2)
```

```python
#练习：阅读字典方法setdefault文档，洗衣歌更简单的invert_dict
def invert_dict(d):
    invert=dict()
    for key in d:
        val=d[key]
        invert[val]=list(invert.setdefault(val,''))+[key]
    return invert
d=histogram('parrot')
invert_dict(d)
```

```python
def invert_dict(d):
    """Inverts a dictionary, returning a map from val to a list of keys.

    If the mapping key->val appears in d, then in the new dictionary
    val maps to a list that includes key.

    d: dict

    Returns: dict
    """
    inverse = {}
    for key,val in d.items():
        inverse.setdefault(val, []).append(key)
    return inverse


if __name__ == '__main__':
    d = dict(a=1, b=2, c=3, z=1)
    inverse = invert_dict(d)
#     for val, keys in inverse.items():
#         print( val, keys)
    print(inverse)
```

```python
#备忘
#斐波那契数列
import time
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
```

```python
start_time=time.time()
fibonacci(30)
end_time=time.time()
jiange_time=end_time-start_time
print(jiange_time)
```

```python
known={0:0,1:1}
def fibonacci1(n):
    if n in known:
        return known[n]
    res=fibonacci1(n-1)+fibonacci1(n-2)
    known[n]=res
    return res
start_time=time.time()
fibonacci1(50)
end_time=time.time()
jiange_time=end_time-start_time
print(jiange_time)
```

```python
fibonacci1(10)
print(known)
```

```python
#练习6-5的备忘形式，注意try后面的情况
cache = {}

def ackermann(m, n):
    """Computes the Ackermann function A(m, n)

    See http://en.wikipedia.org/wiki/Ackermann_function

    n, m: non-negative integers
    """
    if m == 0:
        return n+1
    if n == 0:
        return ackermann(m-1, 1)
    try:
        return cache[m, n]
    except KeyError:
        cache[m, n] = ackermann(m-1, ackermann(m, n-1))
        return cache[m, n]

print (ackermann(3, 4))
print(cache)
# print ackermann(3, 6)
```

```python
#练习：编写一个程序，读入单词表，并找到所有轮转对。轮转对参加（8-12练习中的rotate_word函数）
import bisect
def rotate_word(word,num):
    new_word=''
    for letter in word:
        if letter.isupper():
            start=ord('A')
        elif letter.islower():
            start=ord('a')
        
        new_word=new_word+(chr((ord(letter)-start+num+26)%26+start))
    return new_word
# rotate_word('melon',-10)
def find_lt(a,x):
    i=bisect.bisect_left(a,x)
    if i != len(a) and a[i]==x:
        return True
    return None
def read_words():
    fin=open('words.txt')
    t=[]
    for line in fin:
        word=line.strip()
        t.append(word)
    return t
def sdf():
    t=read_words()
    for s in t:
        for i in range(1,26):
            new_words=rotate_word(s,i)
            if find_lt(t,new_words):
                print(s,i,new_words)
sdf()        
```

```python
def make_word_dict():
    """Read the words in words.txt and return a dictionary
    that contains the words as keys"""
    d = dict()
    fin = open('words.txt')
    for line in fin:
        word = line.strip().lower()
        d[word] = word

    return d
# make_word_dict()
def rotate_pairs(word, word_dict):
    """Prints all words that can be generated by rotating word.

    word: string
    word_dict: dictionary with words as keys
    """
    for i in range(1, 14):
        rotated = rotate_word(word, i)
        if rotated in word_dict:
            print (word, i, rotated)


if __name__ == '__main__':
    word_dict = make_word_dict()

    for word in word_dict:
        rotate_pairs(word, word_dict)
```
