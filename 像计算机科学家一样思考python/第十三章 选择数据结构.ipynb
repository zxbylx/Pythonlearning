```python
def read_file(filename):
    fin=open(filename).readlines()
    for line in fin:
        for i in line:
            if i in string.punctuation: #先将标点符号替换为空格
                line=line.replace(i,' ')
        line.split() #然后再以空格进行单词分拆
        print(line)

read_file('ceshi.txt')
```

```python
#练习：计算书中出现的全部单词总数，以及每个单词的使用次数
def read_file(filename):
    fin=open(filename).readlines()
    count=0
    for line in fin:
        for i in line:
            if i in string.punctuation: #先将标点符号替换为空格
                line=line.replace(i,' ')
        line.split() #然后再以空格进行单词分拆
        count+=1
    return count
read_file('ceshi.txt')
```

```python
def read_file(filename):
    fin=open(filename,encoding='utf-8').read()
    count=0
    for line in fin:
        if line in string.punctuation:
            fin=fin.replace(line,' ')
    x=fin.split()
    d=dict()
    for a in x:
        count+=1
        d[a]=int(d.get(a,'0'))+1
        
    return count,d
read_file('ceshi.txt')
```

```python
#练习：计算书中使用频率最高的20个单词
def read_file(filename):
    fin=open(filename,encoding='utf-8').read()
    count=0
    for line in fin:
        if line in string.punctuation:
            fin=fin.replace(line,' ')
    x=fin.split()
    d=dict()
    for a in x:
        count+=1
        d[a]=int(d.get(a,'0'))+1
        
    return d
def count_top20(d):
    c=[]
    for key,val in d.items():
        c.append((val,key))
    c.sort(reverse=True)
    return c[:20]
a=read_file('powhatan.txt')
count_top20(a)
```
